```python
import random
import hangman_lives as hangman_lives
import hangman_logo
from hangman_words import words

print(hangman_logo.logo)

count=0

find_word=random.choice(words).lower()
print(f"Choosen word is :: {find_word}")
find_word = list(find_word)
word_length=len(find_word)

display=[]
for _ in range(word_length):
  display+="-"

end_of_game = False
print(f"Your Word Consists of these words :: {display}")
while not end_of_game:
  guess = input().lower()
  for i in range(word_length):
    if count >=3:
      print("Game Over You Loose!!!")
      end_of_game = True
      break
    elif guess == find_word[i]:
      display[i]=guess
  print(f"Matched Words ::{display}")
  if guess not in display:
    print(hangman_lives.hangman_lives[count])
    count+=1
    print(f"Remaining Lives ::{4-count}")
  if "-" not in display:
    end_of_game = True
    print("You Win!!!")
```
