```python
from random import choice
from higher_lower_data import data
import higher_lower_logo

print(higher_lower_logo.logo)

def get_data():
  return choice(data)

def compare_age(a,b):
  if a["age"] >= b["age"]:
    return "higher"
  else:
    return "lower"

def game():
  start_game = True
  choice1=get_data()
  result = 0
  while start_game:
    choice2=get_data()
    print(choice1)
    print(choice1["name"])
    print(higher_lower_logo.vers)
    print(choice2)
    print(choice2["name"])

    print("You Guess ? 'higher' or 'lower'")
    user_guess=input().lower()
    original_guess=compare_age(choice1,choice2)
    
    if original_guess == user_guess:
      choice1=choice2
      choice2=get_data()
      result +=1
    else:
      start_game = False
      print(f"Final Score {result}")
    print("-----------------------------------------")

game()
```
